## Load the data and necessary libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt                                                                                                                 

# Load the data
df = pd.read_csv('Dev_data_to_be_shared 3\Dev_data_to_be_shared.csv')
df.head(5)
```

```python
df.info()
```

```python
df = df.fillna(0)
```

```python
df.describe()
```

#####
* since,dataset contains many saprse values and many dimensions.so ,reduce it's dimensions{4} and target attribute is 'bad_flag'.
* drop account_number as it is irrelevant.
* 

```python
df = df.drop(columns=['account_number'],axis=1)
```

```python
df['bad_flag'].value_counts()
```

* since the no.of samples for '1' is low ,as it may lead to model's poor performance and always tend to predict the majority class.
* so,increase the no.of 1's or decrease the no of 0's. perform oversampling and undersampling.
* choosing \* Oversampling as it may increase the no.of instances and make the classes equal. we preferrably use SMOTE{Synthetic Minority Over-Sampling technique}.

### Oversampling

```python
from imblearn.over_sampling import SMOTE

# Assuming your data is in a pandas DataFrame named 'df'
# and the target variable is in the 'target_attribute' column

X = df.drop('bad_flag', axis=1)  # Features
X = X.fillna(0)
y = df['bad_flag']

# Create an instance of SMOTE
smote = SMOTE(random_state=42)

# Perform oversampling
X_resampled, y_resampled = smote.fit_resample(X, y)

# Print the class distribution after oversampling
print(pd.Series(y_resampled).value_counts())
```

```python
X_resampled.shape,y_resampled.shape
```

#### Concatenate the similar columns

```python
col_group = ['onus_attribute_','transaction_attribute_','bureau_enquiry_','bureau_']
df_sum = pd.DataFrame()

for group in col_group:
    
    #select the colummns which startswith similar name
    selected_column = [col for col in X_resampled.columns if col.startswith(group)]

    #sum those values of that columns
    df_sum[group] = X_resampled[selected_column].sum(axis=1)


print(df_sum.shape)
df_sum.head(4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(df_sum.describe())
print(df_sum.info())

# Check for missing values
print(df_sum.isnull().sum())

# Visualizations
# Histogram for each column
```

```python
for col in df_sum.columns:
    plt.figure()
    sns.histplot(df_sum[col], bins=20)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

# Pairplot to visualize relationships between columns
sns.pairplot(df_sum)
plt.show()

# Correlation matrix
correlation_matrix = df_sum.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

### split the data into training and testing & perform scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train , X_test, y_train, y_test = train_test_split(df_sum, y_resampled, test_size=0.25, random_state=41)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

### Model building {perform Ensemble learning}

```python
model1 = RandomForestClassifier(n_estimators=250)
model2 = GradientBoostingClassifier()
model3 = LogisticRegression(penalty='l2',C=0.1)
model4 = SVC()
model5 = DecisionTreeClassifier()
model6 = KNeighborsClassifier(n_neighbors=3)

ensemble_model = VotingClassifier(estimators=[
    ('rf', model1),('lr',model3),('dtclf',model5),('knn',model6)
], voting='hard')

ensemble_model.fit(X_train, y_train)
```

```python
y_pred = ensemble_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Ensemble Model Accuracy:", accuracy)
```

### Metrics

```python
from sklearn.metrics import precision_score,recall_score,f1_score
cfm = confusion_matrix(y_test,y_pred)
prec = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1_s = f1_score(y_test,y_pred)

print('Confusion Matrix:',cfm)
print('Precision:',prec)
print('Recall:',recall)
print('F1 Score:',f1_s)
```
